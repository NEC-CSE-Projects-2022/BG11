Pretraining with Surgical Dataset

```python
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
import pandas as pd

# Load transfer learning dataset
surgical_path = '/content/drive/MyDrive/stroke_prediction/data/Surgical-deepnet.csv'
surgical_df = pd.read_csv(surgical_path)

# One-hot encode and split features/target
surgical_df = pd.get_dummies(surgical_df, drop_first=True)
X_tf = surgical_df.drop('complication', axis=1)
y_tf = surgical_df['complication']

# Impute and scale
X_tf_imputed = IterativeImputer(random_state=0).fit_transform(X_tf)
X_tf_scaled = MinMaxScaler().fit_transform(X_tf_imputed)

# Build base model for pretraining
base_model = Sequential([
    Dense(256, activation='relu', input_dim=X_tf_scaled.shape[1]),
    BatchNormalization(),
    Dropout(0.4),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

base_model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Pretrain the base model
base_model.fit(X_tf_scaled, y_tf, epochs=30, batch_size=32, verbose=1)
```

Code: TL-DNN with 10-Fold Cross-Validation

```python
from sklearn.utils import class_weight
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
import numpy as np
import os

# Ensure save path exists
os.makedirs('/content/drive/MyDrive/stroke_prediction/saved_models', exist_ok=True)

# Initialize Stratified K-Fold
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
acc_list, f1_list, recall_list, precision_list, gmean_list = [], [], [], [], []

# Cross-validation loop
for fold, (train_idx, test_idx) in enumerate(skf.split(X_balanced, y_balanced)):
    print(f"\nðŸ” Fold {fold+1}/10")

    X_train, X_test = X_balanced[train_idx], X_balanced[test_idx]
    y_train, y_test = y_balanced[train_idx], y_balanced[test_idx]

    # Compute class weights
    cw = class_weight.compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
    class_weights_dict = {0: cw[0], 1: cw[1]}

    # Define model (same architecture as base_model)
    model = Sequential([
        Dense(256, activation='relu', input_dim=X_train.shape[1]),
        BatchNormalization(),
        Dropout(0.4),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])

    # Transfer weights from pretrained base_model
    for i in range(1, len(model.layers)):
        model.layers[i].set_weights(base_model.layers[i].get_weights())

    model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])

    # Callbacks
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', patience=4, factor=0.5, verbose=1)
    ]

    # Train model
    history = model.fit(
        X_train, y_train,
        epochs=70,
        batch_size=24,
        validation_split=0.2,
        class_weight=class_weights_dict,
        callbacks=callbacks,
        verbose=1
    )

    # Evaluate
    y_pred_probs = model.predict(X_test).flatten()
    y_pred = (y_pred_probs > 0.5).astype(int)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    gmean = np.sqrt(recall * (cm[0, 0] / (cm[0, 0] + cm[0, 1])))

    print(f"âœ… Fold {fold+1} - Acc: {acc:.4f}, F1: {f1:.4f}, Recall: {recall:.4f}, Precision: {precision:.4f}, G-Mean: {gmean:.4f}")

    # Save results
    acc_list.append(acc)
    f1_list.append(f1)
    recall_list.append(recall)
    precision_list.append(precision)
    gmean_list.append(gmean)

    # âœ… Save the model
    model_path = f'/content/drive/MyDrive/stroke_prediction/saved_models/dnn{fold+1}.keras'
    model.save(model_path)
    print(f"ðŸ’¾ Saved model: dnn{fold+1}.keras")

# âœ… Final average results
print("\nâœ… Final TL-DNN Results:")
print(f"Accuracy : {np.mean(acc_list):.4f}")
print(f"F1 Score : {np.mean(f1_list):.4f}")
print(f"Recall   : {np.mean(recall_list):.4f}")
print(f"Precision: {np.mean(precision_list):.4f}")
print(f"G-Mean   : {np.mean(gmean_list):.4f}")
```

```python
import lime
import lime.lime_tabular
import numpy as np

# Step 1: Define the prediction function for LIME
def predict_proba_fn(x):
    proba = model.predict(x)
    return np.hstack((1 - proba, proba))  # returns [P(No Stroke), P(Stroke)]

# Step 2: Create LIME explainer
lime_explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=X_balanced,
    feature_names=X.columns.tolist(),  # use your actual column names
    class_names=["No Stroke", "Stroke"],
    mode='classification'
)

# Step 3: Choose an instance to explain
i = 10  # you can change this to any index
exp = lime_explainer.explain_instance(
    X_balanced[i],
    predict_proba_fn,
    num_features=10
)

# Step 4: Show the explanation
exp.show_in_notebook()
```
